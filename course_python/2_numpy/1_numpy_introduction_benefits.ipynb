<h3 align="center">Numpy Array vs Python List</h3>

### Benefit # 1: Less Memory

```python
import sys

python_list = list(range(100))
python_list[:4]
```

```python
sys.getsizeof(python_list[0])
```

```python
sys.getsizeof(python_list[0])*len(python_list)
```

```python
numpy_array = np.arange(100)
numpy_array[:4]
```

```python
numpy_array.nbytes
```

As you can see above while numpy array occupies only 800 bytes, the same size list in Python will require 2800 bytes. This shows that numpy arrays are memory efficient

python have non-homogeneous elements. 
In a list in python, elements can be of different data types. For example, a list can contain both integers and strings. So to support this it has to store some extra  meta data And also in list elements are stored in shattered manner. So the total use of memory location of list is more.

But in NumPy  array, elements are stored in contiguous memory location. So it is faster than list. Also it is more memory efficient than list. But it can only store one type of data. For example, it can only store integers or only store strings. So it is more memory efficient than list. But it is less flexible than list.

### Benefit # 2: Speed

```python
import time

SIZE = 1000000

l1 = list(range(SIZE))
l2 = list(range(SIZE))

#to measure the performance of numpy
start_time = time.time()
l3 = [x+y for x, y in zip(l1, l2)]
end_time = time.time()

print("Python list took: ", end_time-start_time)
```

```python
#this code shows about zip() function
l1 = [1, 2, 3]
l2 = [8, 9, 10]
for x,y in zip(l1, l2):
    print(x, y)
```

```python
tuple(zip(l1, l2)) # this will create a list of tuples of pairs
```

```python
n1 = np.arange(SIZE)
n2 = np.arange(SIZE)
n1[:3], n2[:3]
```

```python
start_time = time.time()
n3 = n1 + n2
end_time = time.time()
print("Numpy array took: ", end_time-start_time)
```
