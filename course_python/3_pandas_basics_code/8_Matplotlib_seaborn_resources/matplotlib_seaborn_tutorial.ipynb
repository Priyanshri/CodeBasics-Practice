```python
import pandas as pd

df_sales = pd.read_excel("linechart.xlsx")
df_sales.head()
```

## Line Chart

```python
from matplotlib import pyplot as plt
plt.plot(df_sales['Quarter'], df_sales['Fridge'])
# this will plot the line chart for Fridge sales
```

```python


plt.figure(figsize=(12,4)) # Set the figure size

plt.plot(df_sales["Quarter"], df_sales["Fridge"], color="blue", label="Fridge")
plt.plot(df_sales["Quarter"], df_sales["Dishwasher"], color="orange", label="Dishwasher")
plt.plot(df_sales["Quarter"], df_sales["Washing Machine"], color="green", label="Washing Machine")

plt.title("Product Sales") # Set the title of the plot
plt.xlabel("Quarter") # Set the x-axis label
plt.ylabel("Million $") # Set the y-axis label

plt.legend() # Show the legend showing which line corresponds to which product

plt.show()
```

## Pie Chart

```python
total_sales = df_sales[["Fridge","Dishwasher","Washing Machine"]].sum()
# this will sum the sales of all products across all quarters
total_sales
```

```python
plt.pie(total_sales, labels= total_sales.index, autopct='%1.2f%%', startangle= 140, explode=(0.1,0,0), shadow=True)
# this will plot the pie chart for total sales of all products
#here labels shows the names of the products, autopct shows the percentage of each product in the total sales, startangle (in degree) rotates the pie chart, explode separates the first slice from the rest, and shadow adds a shadow effect to the pie chart 
plt.show()
```

## Bar Chart

```python
df_sales.plot(kind="bar", x="Quarter") # this will plot the bar chart for sales of all products across all quarters 
plt.xticks(rotation=45) # Rotate x-axis labels for better readability
plt.show()
```

To use Quarters as X axis, we can set them as an index on a dataframe and then use plot function

```python
df_sales_2 = df_sales.set_index("Quarter")
# this will set the Quarter column as the index of the DataFrame
df_sales_2
```

```python
df_sales_2.loc["Q1 2022"] 
# this will return the sales data for [ Q1 2022 ]
```

```python
df_sales_2.plot(kind="bar")
plt.xticks(rotation=45)
plt.show()
```

```python
df_sales_2.plot(kind="barh")
plt.show()
```

## Histogram

```python
df_score = pd.read_excel("histograms.xlsx")
df_score.head(3)
```

```python
plt.hist(df_score["Exam_Score"], density=True)
# this will plot the histogram for Exam_Score and show the distribution of scores
```

```python
import seaborn as sns
sns.histplot(df_score["Exam_Score"])
# this will plot the histogram for Exam_Score and show the distribution of scores
```

```python
sns.histplot(df_score["Exam_Score"], kde=True)
# this will plot the histogram for Exam_Score and show the distribution of scores with a kernel density estimate (KDE) line
```

```python
if 'df_score' not in locals():
	df_score = pd.read_excel("histograms.xlsx")
# Load the DataFrame if it doesn't exist

sns.histplot(df_score["Exam_Score"], kde=True) 

plt.xlabel("Score")
plt.ylabel("Count")
plt.title("Exam Scores")
plt.show()
```
