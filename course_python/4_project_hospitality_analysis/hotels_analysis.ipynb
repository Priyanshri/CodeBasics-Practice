### ==> 2. Data Cleaning
***

```python
df_bookings.describe()
```

**(1) Clean invalid guests**

```python
df_bookings[df_bookings.no_guests<=0]
# This will show you bookings with no. of guests less than 0, which should not happen 
```

```python
df_bookings.shape #dataframe shape before filtering
```

As you can see above, number of guests having less than zero value represents data error. We can ignore these records.

```python
df_bookings = df_bookings[df_bookings.no_guests>0]
# This will filter out bookings with no. of guests less than 0 and keep only valid bookings and store it back in df_bookings
```

```python
df_bookings.shape #dataframe shape after filtering
```

**(2) Outlier removal in revenue generated**

```python
df_bookings.revenue_generated.min(), df_bookings.revenue_generated.max()
# This will give you the minimum and maximum revenue generated from bookings which has some data error
```

```python
df_bookings.revenue_generated.mean(), df_bookings.revenue_generated.median()
# This will calculate the average and median of the revenue generated
```

Sandard deviation is based on this image.

```python
avg, std = df_bookings.revenue_generated.mean(), df_bookings.revenue_generated.std()
# This will calculate the average and standard deviation of the revenue generated 
avg, std
```

```python
higher_limit = avg + 3*std 
# This will calculate the higher limit for outliers based on the average and standard deviation
higher_limit
```

```python
lower_limit = avg - 3*std
# This will calculate the lower limit for outliers based on the average and standard deviation
lower_limit
```

Now we are checking if revenue_generated is less than 0 and greater than higher limit or not because revenue can neither be less than 0 nor greater than higher limit value.

```python
df_bookings[df_bookings.revenue_generated<=0]
# This will show you bookings with revenue generated less than or equal to 0, which should not happen
```

```python
df_bookings[df_bookings.revenue_generated>higher_limit]
# This will show you bookings with revenue generated greater than the higher limit, which are considered outliers
```

```python
df_bookings = df_bookings[df_bookings.revenue_generated<=higher_limit]
#cleaning the data by removing outliers based on the higher limit
df_bookings.shape
```

### ==> 3. Data Transformation
***

**Data Transformation** means you have your data and you are transforming to a state such that you can perform your anlytics better

**Create occupancy percentage column**

```python
df_agg_bookings.head(3)
```

```python
#calculating occupancy percentage i.e. here, successful_bookings/capacity
25/30
```

```python
df_agg_bookings['occ_pct'] = df_agg_bookings.apply(lambda row: row['successful_bookings']/row['capacity'], axis=1)
# This will calculate the occupancy percentage for each row in df_agg_bookings and add it as a new column 'occ_pct'

df_agg_bookings.head(3)
```

```python
# Multiply the existing numeric occ_pct (fraction) by 100 and round to 2 decimals.
# Use vectorized operations on the occ_pct series instead of DataFrame.apply on rows.
df_agg_bookings['occ_pct'] = (df_agg_bookings['occ_pct'].astype(float) * 100).round(2)
df_agg_bookings.head(3)
```

**Forms of Data Transformation**









You can use following approach to get rid of SettingWithCopyWarning

```python
new_col = df_agg_bookings.apply(lambda row: row['successful_bookings']/row['capacity'], axis=1)
df_agg_bookings = df_agg_bookings.assign(occ_pct=new_col.values)
df_agg_bookings.head(3)
```

Convert it to a percentage value

```python
df_agg_bookings['occ_pct'] = df_agg_bookings['occ_pct'].apply(lambda x: round(x*100, 2))
df_agg_bookings.head(3)
```
